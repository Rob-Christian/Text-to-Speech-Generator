```python
# Import necessary libraries
from transformers import pipeline
import torch
from IPython.display import Audio
```

```python
# Define function for text-to-speech
def text_to_speech(text):
  # Load the pipeline
  tts_pipeline = pipeline("text-to-speech", model = "suno/bark-small", device = "cuda" if torch.cuda.is_available() else 'cpu')

  # Generate the speech
  print("Generating Speech...")
  output = tts_pipeline(text)

  # Play audio
  print("Playing the audio...")
  return Audio(output['audio'], rate = output["sampling_rate"])
```

```python
# Main program
if __name__ == "__main__":
  text = input("Enter the text you want to convert to speech: ")
  audio_output = text_to_speech(text)
  display(audio_output)
```
